```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='ISO-8859-1')
```

```python
df.head()
```

```python
# Recommended way
# fit_reg=False remove the regression line
# hue='Stage' point pokemon revolution stage
sns.lmplot(x='Attack', y='Defense', data=df,
          fit_reg=False,
          hue='Stage')

# Tweak with matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
# Pre-format DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

# Boxplot
sns.boxplot(data=stats_df)

# Pandas cheat sheet
# http://pandas.pydata.org/pandas-docs/stable/10min.html
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]

# set theme
sns.set_style('whitegrid')

# violin plot
sns.violinplot(x='Type 1', y='Attack', data=df,
              palette=pkmn_type_colors)
```

```python
# swarm plot with pokemon color palette
sns.swarmplot(x='Type 1', y='Attack', data=df,
             palette=pkmn_type_colors)
```

```python
# Set figure size with matplotlib
plt.figure(figsize=(10,6))
 
# Create plot
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Attack by Type')
```

```python
stats_df.head()
```

```python
# Melt DataFrame
melted_df = pd.melt(stats_df,
                   id_vars=['Name', 'Type 1', 'Type 2'],
                   var_name='Stat')
melted_df.head()
```

```python
print(stats_df.shape)
print(melted_df.shape)
```

```python
# swarmplot melted_df
sns.swarmplot(x='Stat', y='value', data=melted_df,
             hue='Type 1')
```

```python
# 1. Enlarge the plot
plt.figure(figsize=(10,6))
 
sns.swarmplot(x='Stat', 
              y='value', 
              data=melted_df, 
              hue='Type 1', 
              dodge=True, # 2. Separate points by hue
              palette=pkmn_type_colors) # 3. Use Pokemon palette
 
# 4. Adjust the y-axis
plt.ylim(0, 260)
 
# 5. Place legend to the right
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

```python
# Calculate correlations
corr = stats_df.corr()
 
# Heatmap
sns.heatmap(corr)
```

```python
# Distribution Plot (a.k.a. Histogram)
sns.distplot(df.Attack)
```
